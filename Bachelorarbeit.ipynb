```python
from google.genai import types

MODEL_ID = "gemini-3-pro-preview"

PROMPTS = [
    "Fasse das Video in maximal drei Sätzen zusammen.",
    "Fasse das Video in Stichpunkten zusammen.",
    "Welche Hauptinhalte sind im Video zu sehen?",
    "Welche wichtigen Ereignisse sind im Video zu sehen?",
    "Welche Informationen fehlen oder bleiben unklar?",
]

def ask_gemini_about_youtube(youtube_url: str, prompt: str) -> str:
    contents = [
        types.Content(
            parts=[
                types.Part.from_uri(file_uri=youtube_url, mime_type="video/*"),
                types.Part(text=prompt),
            ]
        )
    ]
    resp = client.models.generate_content(model=MODEL_ID, contents=contents)
    return resp.text

youtube_url = input("YouTube-Video-URL einfügen: ").strip()

for i, p in enumerate(PROMPTS, start=1):
    print(f"\n--- Prompt {i}/{len(PROMPTS)} ---")
    print("FRAGE:", p)
    try:
        ans = ask_gemini_about_youtube(youtube_url, p)
        print("ANTWORT:\n", ans)
    except Exception as e:
        print("FEHLER:", repr(e))
```
